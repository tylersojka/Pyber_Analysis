```python
# make sure our plots are displayed inline within the Jupyter Notebook directly below the code cell that produced it
%matplotlib inline

# Import dependencies
import matplotlib.pyplot as plt

import numpy as np
```

```python
# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
# Create the plot
plt.plot(x_axis, y_axis)
```

```python
# Create the plot with ax.plt()
fig, ax = plt.subplots()
ax.plot(x_axis, y_axis)
```

```python
# Create the plot with ax.plt()
ax = plt.axes()
ax.plot(x_axis, y_axis)
```

```python
# Create the plot.
plt.plot(x_axis, y_axis)
plt.show()
```

```python
# Create the plot and add a label for the legend.
plt.plot(x_axis, y_axis, label='Boston')
# Create labels for the x and y axes.
plt.xlabel("Date")
plt.ylabel("Fare($)")
# Set the y limit between 0 and 45.
plt.ylim(0, 45)
# Create a title.
plt.title("PyBer Fare by Month")
# Add the legend.
plt.legend()
```

```python
# Create the plot.
plt.plot(x_axis, y_axis, marker="*", color="blue", linewidth=2, label='Boston')
# Create labels for the x and y axes.
plt.xlabel("Date")
plt.ylabel("Fare($)")
# Set the y limit between 0 and 45.
plt.ylim(0, 45)
# Create a title.
plt.title("PyBer Fare by Month")
# Add a grid.
plt.grid()
# Add the legend.
plt.legend()
```

```python
ax = plt.axes()
ax.plot(x_axis, y_axis, marker="d", color="green", linewidth=2, label="boston")
ax.set_ylim(0, 45)
ax.legend()
ax.set_title("PyBer Fare by Month")
ax.grid()
ax.set_ylabel("Fares($)")
ax.set_xlabel("Date")
```

```python
# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = np.array([10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09])
```

```python
#using the plt.plot() function
plt.plot(x_axis, y_axis, 'o')
```

```python
# using the plt.scatter() function
plt.scatter(y_axis, x_axis, c="r", label="Chigago")
plt.xlabel("Fare($)")
plt.ylabel("Date")
plt.title("PyBer Fare by Month")
plt.legend()
plt.gca().invert_yaxis()
```

```python
# bubble plot, size set to y axis for dif size
plt.scatter(x_axis, y_axis, s=y_axis)
```

```python
y_axis_larger = []
for data in y_axis:
  y_axis_larger.append(data*3)
plt.scatter(x_axis, y_axis, s=y_axis_larger)
```

```python
# list comprehension to replace the for loop to increase y values
plt.scatter(x_axis, y_axis, s = [i * 3 for i in y_axis])
```

```python
# scatter plot using object oriented method
fig, ax = plt.subplots()
ax.scatter(x_axis, y_axis)
```

```python
fig, ax = plt.subplots()
ax.scatter(x_axis, y_axis, s= [i * 5 for i in y_axis], label="Boston", c="skyblue", linewidth=2, alpha=.35, edgecolors="black")
```

```python
fig, ax = plt.subplots()
ax.scatter(y_axis, x_axis, s= [i * 10 for i in y_axis], label="Boston", color="skyblue", linewidth=2, alpha=.80, edgecolors="black")
plt.gca().invert_yaxis()
ax.set_ylabel("Dates")
ax.set_xlabel("Fares($)")
ax.set_title("PyBer Fares by Month")
ax.set_xlim(0, 45)
ax.legend()
```

```python
plt.subplots(figsize=(8, 8))
explode_values = (0, 0, 0, 0, 0, 0, 0.15, 0, 0, 0, 0, 0)
plt.pie(y_axis, explode=explode_values, labels=x_axis, autopct='%.1f%%')
plt.show()
```
